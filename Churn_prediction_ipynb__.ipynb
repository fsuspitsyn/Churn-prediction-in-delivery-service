```python
# Create a copy of data and map platform codes to names
df_platforms = df.copy()
df_platforms['pl'] = df_platforms.pl.round()
platform_mapping = {1: 'Android', 2: 'iOS', 4: 'WEB'}
df_platforms['platform_name'] = df_platforms['pl'].map(platform_mapping)

# Group by customer and platform
platform_churn = df_platforms.groupby(['cid', 'platform_name'])['churn'].max().reset_index()

# Filter out platforms with zero customers
platform_churn = platform_churn[platform_churn.groupby('platform_name')['platform_name'].transform('count') > 0]

# Get platform order by popularity
platform_order = platform_churn['platform_name'].value_counts().index

# Create plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='platform_name', hue='churn', data=platform_churn,
                  order=platform_order, palette=['#1f77b4', '#ff7f0e'])

# Calculate totals per platform
platform_totals = platform_churn['platform_name'].value_counts().loc[platform_order]

# Add percentage annotations (only for non-zero bars)
for i, p in enumerate(ax.patches):
    if p.get_height() > 0:  # Only label non-zero bars
        platform_idx = i % len(platform_order)
        platform = platform_order[platform_idx]
        total = platform_totals[platform]
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height() + 0.03 * max(platform_totals)
        ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10)

plt.title('Customer Churn by Platform', pad=20, fontsize=14)
plt.xlabel('Platform', labelpad=10, fontsize=12)
plt.ylabel('Number of Customers', labelpad=10, fontsize=12)
plt.legend(title='Status', labels=['Retained (0)', 'Churned (1)'],
           bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylim(0, max([p.get_height() for p in ax.patches]) * 1.15)  # Add 15% padding
plt.xticks(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Копируем, чтобы не портить оригинал
df = df.copy()

# 1. Считаем, сколько разных месяцев активен клиент
months_active = df.groupby('cid')['mnth_n']\
                  .nunique()\
                  .rename('months_active')

# 2. Добавляем censored = 1 для клиентов с активностью < 3 месяцев
censored = (months_active < 3).astype(int).rename('censored')

# 3. Находим для каждого клиента, в каком месяце он помечен churn==1 (если есть)
churn_month = df.loc[df['churn'] == 1]\
                .groupby('cid')['mnth_n']\
                .min()\
                .rename('churn_month')

# Вставляем эти серии в единый DataFrame-«шаблон»
agg = pd.concat([months_active, censored, churn_month], axis=1)

# 4. Присоединяем к каждому событию номер его churn_month (для фильтраций внутри окон)
df = df.merge(churn_month, on='cid', how='left')

# 5. Отмечаем строки в «окнах» до оттока
df['in_30d_before']    = df['mnth_n'] == df['churn_month'] - 1
df['in_60_90d_before'] = df['mnth_n'].between(df['churn_month'] - 3,
                                              df['churn_month'] - 2)

# 6. Основные агрегаты по клику
group = df.groupby('cid')

# Набор фичей
features = pd.DataFrame(index=group.size().index)

# — количество платформ, городов, магазинов и месяцев с покупками
features['n_platforms']    = group['pl'].nunique()
features['n_cities']       = group['gcid'].nunique()
features['n_shops']        = group['shpid'].nunique()
features['months_active']  = months_active

# — покупки в окна до оттока
features['cnt_30d_before']    = group.apply(
    lambda g: g.loc[g['in_30d_before'],    'cnt_cpl'].sum()
)
features['cnt_60_90d_before'] = group.apply(
    lambda g: g.loc[g['in_60_90d_before'], 'cnt_cpl'].sum()
)

# — средний чек в тех же окнах
def avg_check(sub: pd.DataFrame):
    total_sum = sub['sum_inv'].sum()
    total_cnt = sub['cnt_cpl'].sum()
    return total_sum / total_cnt if total_cnt > 0 else 0

features['avg_check_30d']    = group.apply(
    lambda g: avg_check(g.loc[g['in_30d_before']])
)
features['avg_check_60_90d'] = group.apply(
    lambda g: avg_check(g.loc[g['in_60_90d_before']])
)

# — просто средний чек за весь период
features['avg_check_all'] = group.apply(avg_check)

# — офлайн-покупки
features['ever_offline']   = (group['cnt_of'].sum() > 0).astype(int)
features['cnt_offline']    = group['cnt_of'].sum().fillna(0)

# — промокоды, отмены, фрод
features['cnt_promo']      = group['sm_pr'].sum().fillna(0)
features['cnt_cancel']     = group['cnt_cnl'].sum().fillna(0)
features['cnt_fraud']      = group['cnt_fr'].sum().fillna(0)

# — среднее расстояние до магазина (без самовывоза) и бинning
dist = df.loc[(df['dstnc']>0) & (df['dlvr_md']!='pckp')]
avg_dist = dist.groupby('cid')['dstnc'].mean().rename('avg_dist')
bins = [0, 500, 1000, 2000, 5000, np.inf]
# labels = ['<=500m','500-1km','1-2km','2-5km','>5km']
labels = [0, 0.5, 1, 2, 5]
dist_bin = pd.cut(avg_dist, bins=bins, labels=labels).rename('dist_bin')
features = features.join(avg_dist).join(dist_bin)

# — прочие агрегаты
features['max_orders_series'] = group['max_ser'].max().fillna(0)
features['sum_inv_delivery']  = df.loc[df['dlvr_md']=='crr']\
                                   .groupby('cid')['sum_inv'].sum()
features['sum_inv_pickup']    = df.loc[df['dlvr_md']=='pckp']\
                                   .groupby('cid')['sum_inv'].sum()

# — время с последнего заказа (предположим, что всего 12 месяцев)
features['months_since_last'] = 12 - group['mnth_n'].max()

# # — интервалы между месяцами (по уникальным)
# def gaps(series):
#     us = np.sort(series.unique())
#     if len(us) < 2:
#         return pd.Series({'max_gap':0, 'mean_gap':0})
#     diffs = np.diff(us)
#     return pd.Series({'max_gap':diffs.max(), 'mean_gap':diffs.mean()})

# g = df.groupby('cid')['mnth_n'].apply(gaps)
# features = features.join(g)

# — предпочитаемые значения (mode)
features['pref_city']     = group['gcid'].agg(lambda x: x.mode().iloc[0])
features['pref_shop']     = group['shpid'].agg(lambda x: x.mode().iloc[0])
features['pref_platform'] = group['pl'].agg(lambda x: x.mode().iloc[0])
```

```python
# # — среднее кол-во товаров в заказе
# tot = group[['sm_it','cnt_cpl']].sum()
# features['avg_items_per_order'] = np.where(
#     tot['cnt_cpl']>0,
#     tot['sm_it'] / tot['cnt_cpl'],
#     0
# )

# Убираем всю предыдущую логику с tot = … и np.where
# Вместо этого просто:
features['avg_items_per_order'] = df.groupby('cid').apply(
    lambda g: g['sm_it'].sum() / g['cnt_cpl'].sum()
              if g['cnt_cpl'].sum() > 0 else 0
)


# — итоговые метки: churn, censored и неагрегируемые фичи
final = features.copy()
final['churn']   = group['churn'].max()
final['censored']= censored
# неагрегируемые (берём первое значение)
final['gender']      = group['gender'].first()
final['age']         = group['ag'].first()
final['days_first']  = group['c_frst'].first()

# Наконец, итоговый DataFrame
data = final.reset_index()
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Шаг 1: Подготовка данных

features_list = ['n_platforms', 'n_cities', 'n_shops',
       'avg_check_all', 'ever_offline',
       'cnt_promo', 'cnt_cancel', 'cnt_fraud', 'avg_dist', 'dist_bin',
        'sum_inv_delivery', 'sum_inv_pickup',
        'pref_city', 'pref_shop', 'pref_platform',
       'avg_items_per_order', 'gender', 'age', 'age_segment',
        'avg_price_item', 'cnt_offline', 'first_with_promo', 'days_first', 'cnt_pckp']

# Признаки (X) и целевая переменная (y)
X = data.query('days_first >= 0')[features_list]  # Используем cid, mnth_n и sum_inv как признаки
y = data.query('days_first >= 0')[['churn']]

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Преобразуем данные в numpy.array
X_train = X_train.to_numpy()
X_test = X_test.to_numpy()
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()

# Шаг 2: Обучение модели XGBoost
model = XGBClassifier(random_state=42, eval_metric='logloss')
model.fit(X_train, y_train)

# Шаг 3: Предсказание и оценка модели
y_pred = model.predict(X_test)  # Предсказанные значения
y_pred_proba = model.predict_proba(X_test)[:, 1]  # Вероятности для ROC-AUC

# Метрики качества
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Матрица
conf_matrix = confusion_matrix(y_test, y_pred)

# Вывод результатов
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"ROC-AUC: {roc_auc:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, StackingClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, precision_recall_curve, auc
)

# ----------------------------------------
# 1) Загрузка и подготовка данных
# ----------------------------------------
# df = pd.read_csv('your_data.csv')
# features = df[features_list]
# target   = df['churn']
# X_train, X_test, y_train, y_test = train_test_split(
#     features, target, test_size=0.3, random_state=42
# )

# Приводим y к одномерному виду
# y_train = y_train.values.ravel()
# y_test  = y_test.values.ravel()

# ----------------------------------------
# 2) Инициализация моделей
# ----------------------------------------
models = {
   # "Logistic Regression": LogisticRegression(max_iter=2000, random_state=42),
    "Random Forest":       RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost":             XGBClassifier(eval_metric="logloss", use_label_encoder=False, random_state=42),
    "CatBoost":            CatBoostClassifier(verbose=0, random_state=42),
    "LightGBM":            LGBMClassifier(random_state=42),
    "MLP":                 MLPClassifier(hidden_layer_sizes=(50,30), max_iter=500, random_state=42)
}

# ----------------------------------------
# 3) Обучение базовых моделей и сбор метрик
# ----------------------------------------
results   = []
roc_curves = {}
pr_curves  = {}

for name, clf in models.items():
    t0 = time.time()
    clf.fit(X_train, y_train)
    train_time = time.time() - t0

    y_pred = clf.predict(X_test)
    y_proba = clf.predict_proba(X_test)[:, 1]

    # считаем метрики
    acc     = accuracy_score(y_test, y_pred)
    prec    = precision_score(y_test, y_pred)
    rec     = recall_score(y_test, y_pred)
    f1      = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)

    results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1": f1,
        "ROC-AUC": roc_auc,
        "Train Time (s)": round(train_time, 2)
    })

    # строим ROC-кривую
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_curves[name] = (fpr, tpr, roc_auc)

    # строим Precision-Recall кривую
    prec_vals, rec_vals, _ = precision_recall_curve(y_test, y_proba)
    pr_auc = auc(rec_vals, prec_vals)
    pr_curves[name] = (rec_vals, prec_vals, pr_auc)

# ----------------------------------------
# 4) Ансамбли: Voting и Stacking
# ----------------------------------------
# Voting (soft voting)
voters = [(n, models[n]) for n in ["XGBoost", "CatBoost", "LightGBM", "MLP"]]
voting = VotingClassifier(voters, voting='soft', weights=[2,2,2,1], n_jobs=-1)
t0 = time.time()
voting.fit(X_train, y_train)
v_time = time.time() - t0

y_pred_v = voting.predict(X_test)
y_proba_v = voting.predict_proba(X_test)[:,1]
results.append({
    "Model": "Voting",
    "Accuracy": accuracy_score(y_test, y_pred_v),
    "Precision": precision_score(y_test, y_pred_v),
    "Recall": recall_score(y_test, y_pred_v),
    "F1": f1_score(y_test, y_pred_v),
    "ROC-AUC": roc_auc_score(y_test, y_proba_v),
    "Train Time (s)": round(v_time, 2)
})
fpr, tpr, _ = roc_curve(y_test, y_proba_v)
roc_curves["Voting"] = (fpr, tpr, roc_auc_score(y_test, y_proba_v))
prec_vals, rec_vals, _ = precision_recall_curve(y_test, y_proba_v)
pr_curves["Voting"] = (rec_vals, prec_vals, auc(rec_vals, prec_vals))

# Stacking
stack = StackingClassifier(estimators=voters, final_estimator=XGBClassifier(), passthrough=True, n_jobs=-1)
t0 = time.time()
stack.fit(X_train, y_train)
s_time = time.time() - t0

y_pred_s = stack.predict(X_test)
y_proba_s = stack.predict_proba(X_test)[:,1]
results.append({
    "Model": "Stacking",
    "Accuracy": accuracy_score(y_test, y_pred_s),
    "Precision": precision_score(y_test, y_pred_s),
    "Recall": recall_score(y_test, y_pred_s),
    "F1": f1_score(y_test, y_pred_s),
    "ROC-AUC": roc_auc_score(y_test, y_proba_s),
    "Train Time (s)": round(s_time, 2)
})
fpr, tpr, _ = roc_curve(y_test, y_proba_s)
roc_curves["Stacking"] = (fpr, tpr, roc_auc_score(y_test, y_proba_s))
prec_vals, rec_vals, _ = precision_recall_curve(y_test, y_proba_s)
pr_curves["Stacking"] = (rec_vals, prec_vals, auc(rec_vals, prec_vals))

# ----------------------------------------
# 5) Вывод итогов и визуализация
# ----------------------------------------
comparison_df = pd.DataFrame(results).sort_values("ROC-AUC", ascending=False)
print(comparison_df)

# ROC-кривые
plt.figure(figsize=(10,6))
for name, (fpr, tpr, auc_val) in roc_curves.items():
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc_val:.2f})")
plt.plot([0,1],[0,1],'k--',alpha=0.5)
plt.title("ROC Curves")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Precision-Recall кривые
plt.figure(figsize=(10,6))
for name, (rec_vals, prec_vals, pr_auc) in pr_curves.items():
    plt.plot(rec_vals, prec_vals, label=f"{name} (AUC={pr_auc:.2f})")
plt.title("Precision-Recall Curves")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import optuna
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
import numpy as np

def objective(trial):
    # 1) Определяем пространство гиперпараметров
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'max_depth':    trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.3),
        'subsample':     trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'gamma':         trial.suggest_float('gamma', 0, 5),
        'reg_alpha':     trial.suggest_float('reg_alpha', 0, 5),
        'reg_lambda':    trial.suggest_float('reg_lambda', 0, 5),
        'use_label_encoder': False,
        'eval_metric': 'logloss',
        'random_state': 42
    }

    # 2) Создаём модель с этими гиперпараметрами
    model = XGBClassifier(**param)

    # 3) Кросс-валидация
    cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='roc_auc', n_jobs=-1)

    # 4) Optuna минимизирует, поэтому возвращаем −mean(AUC)
    return 1.0 - np.mean(scores)

# 5) Создаём и запускаем исследование
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50, timeout=600)  # либо по числу итераций, либо по времени

# 6) Вывод лучших параметров
print("Best trial:")
trial = study.best_trial
print(f"  AUC: {1-trial.value:.4f}")
print("  Params:")
for key, val in trial.params.items():
    print(f"    {key}: {val}")
```

```python
import json

# 7) Сохраняем лучшие параметры в файл params.json
with open('params.json', 'w') as f:
    json.dump(trial.params, f, indent=4)

print("Best parameters saved to 'params.json'")
```
